```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self):
        self.a = 0
        self.b = 0

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        self.a = np.cov(X, y,bias=True)[0, 1] / np.var(X)
        self.b = np.mean(y) - self.a * np.mean(X)

    def predict(self, X):
        predictions = []
        for x in X:
            predictions.append(self.a * x + self.b)
        return predictions
```

```python
X = [1, 2, 3, 4]
y = [2, 4, 6, 8]

model = LinearRegression()
model.fit(X, y)

print(model.predict([5]))
```
