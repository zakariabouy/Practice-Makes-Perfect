```python
import numpy as np
```

```python
class LogisticRegressionGD:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.w = 0
        self.b = 0

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        n = len(y)

        self.w = 0
        self.b = 0

        for _ in range(self.n_iters):
            y_pred = self.sigmoid(self.w * X + self.b)

            dw = (1/n) * np.sum(X * (y_pred - y))
            db = (1/n) * np.sum(y_pred - y)

            self.w -= self.lr * dw
            self.b -= self.lr * db

    def predict_proba(self, X):
        X = np.array(X)
        return self.sigmoid(self.w * X + self.b)

    def predict(self, X):
        probs = self.predict_proba(X)
        return (probs >= 0.5).astype(int)
```

```python
X = [1,2,3,4,5,6]
y = [0,0,0,1,1,1]
```

```python
model = LogisticRegressionGD(learning_rate=0.01, n_iters=5000)
model.fit(X,y)
print(model.predict_proba([1,6]))
print(model.predict([1,6]))
```
