```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self):
        self.a = 0
        self.b = 0
        
    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        
        self.a = np.cov(X, y,bias=True)[0, 1] / np.var(X)
        self.b = np.mean(y) - self.a * np.mean(X)

    def predict(self, X):
        self.predictions = []
        for x in X:
            self.predictions.append(self.a * x + self.b)
        return self.predictions
    
    def R_squared(self, X, y):
        X = np.array(X)
        y = np.array(y)
        predictions = np.array(self.predict(X))
        Sr = np.sum((y - predictions) ** 2)
        St = np.sum((y - np.mean(y)) ** 2)
        return 1 - Sr / St
```

```python
X = [1, 2, 3, 4]
y = [2, 4, 6, 8]

model = LinearRegression()
model.fit(X, y)

print(model.predict([5]))
```

```python
X = [1,2,3,4]
y = [2,4,6,8]

model = LinearRegression()
model.fit(X,y)

print(model.score(X,y))
```

```python
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.a = 0
        self.b = 0
        
    def fit(self, X, y):
        X=np.array(X)
        y=np.array(y)
        self.a=0
        self.b=0
        n=len(y)
        for _ in range(self.n_iters):
            y_pred = self.a*X+self.b
            da = (-2/n)*np.sum(X*(y-y_pred))
            db = (-2/n)*np.sum(y-y_pred)
            self.a -= da*self.lr
            self.b -= db*self.lr
    
    def predict(self, X):
        X=np.array
        
        return self.a*X+self.b
```

```python
X=[1,2,3,4]
y=[2,4,6,8]

model= LinearRegressionGD()
model.fit(X,y)
print(model.predict([5]))
```
